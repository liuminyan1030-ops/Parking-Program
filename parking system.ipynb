```python
from datetime import datetime

# ===================== Constants =====================
MAXIMUM_CAPACITY = 50
PARKING_FEE = 4
ADMIN_PASSWORD = "password"

# ===================== Global Lists =====================
plates = []
cc_numbers = []
parking_space = 0

# ===================== Functions =====================
# Register a vehicle if there is available space.
def register_vehicle():
    global parking_space
    if parking_space < MAXIMUM_CAPACITY:
        print("The parking lot has spaces for your vehicle")
        license_number = input("Enter your plate number: ").strip().upper()
        if license_number not in plates:
            credit_card_number = input("Enter your Credit Card Number ($4.00 charge): ").strip()
            cc_numbers.append(credit_card_number)
            plates.append(license_number)
            parking_space += 1
            print(f"Thank you, your plate {license_number} has been added to the lot.")
        else:
            print(f'{license_number} is already registered')
    else:
        print("The parking lot is full")
# Verify if a vehicle is in the parking lot.
def verify_vehicle():
    license_number = input("Enter your plate number to verify: ")
    if license_number in plates:
        print(f'The vehicle with plate# {license_number} is registered in the lot')
    else:
        print(f'The vehicle is NOT registered')
#Display registered vehicles and save them to a file.
def display_vehicles():
    current_time = datetime.now().strftime("%Y-%m-%d")
    if not plates:
        print("The parking lot is empty!")
    else:
        print(f"A list of registered vehicles for {current_time}")
        print("="*30)
        print(f"{'Plate':^20}")
        print("="*30)
        with open("vehicles.txt", "w") as f:
            for plate in plates:
                print(f'{plate:^20}')
                f.write(plate + "\n")
#Display daily charges and save them to a file.
def display_charges():
    current_time = datetime.now().strftime("%Y-%m-%d")
    if not plates:
        print("The parking lot is empty!")
    else:
        print(f"Daily parking summary for {current_time}")
        print("="*60)
        print(f"{'Plate':>12} {'Credit Card':>30} {'Charge in $':>15}")
        print("="*60)
        total_charge = 0
        with open("charges.txt", "w") as f:
            for i in range(len(plates)):
                plate = plates[i]
                cc = cc_numbers[i]
                print(f'{plate:>12} {cc:>30} {PARKING_FEE:>15}')
                f.write(f"{plate}\t{cc}\t{PARKING_FEE:.2f}\n")
                total_charge += PARKING_FEE
        print("="*60)
        print(f'{"Total":>12} {" ":>30} {total_charge:>15}')
#Remove a specific vehicle from the lot.
def remove_vehicle():
    global parking_space
    license_number = input("Enter the plate number to remove: ")
    if license_number in plates:
        index = plates.index(license_number)
        plates.pop(index)
        cc_numbers.pop(index)
        parking_space -= 1
        print(f'{license_number} is removed')
    else:
        print(f"{license_number} is not registered")
# Clear all vehicles from the lot.
def clear_vehicles():
    global parking_space
    plates.clear()
    cc_numbers.clear()
    parking_space = 0
    print("All vehicles were removed and the lot is empty")
# Check admin password before accessing restricted options.
def check_password():
    password = input("Enter your password: ")
    if password == ADMIN_PASSWORD:
        return True
    else:
        print("Password is incorrect!")
        return False
#Display the main menu and process user choices.
def print_menu():
    choice = -1
    while choice != 0:
        print("*"*80)
        print("*** Welcome to Park and Go Parking Application!***")
        print("Park from 6 PM- Midnight for a flat fee of $4.00 ")
        print("*"*80)
        print("Select from the following options")
        print("1-Register a vehicle")
        print("2-Verify vehicle registration")
        print("3-Display registered vehicles and save them to a file")
        print("4-Display daily charges and save them to a file ")
        print("5-Remove a vehicle")
        print("6-Clear  vehicles")
        print("0-Exit")

        user_input = input(">>> ").strip()
        if user_input.isdigit():
            choice = int(user_input)
            admin_options = [2,3,4,5,6]

            if choice == 2:
                print("Verify your registration")
                if check_password():
                    verify_vehicle()
            elif choice in [3,4,5,6]:
                if check_password():
                    if choice==3:
                      display_vehicles()
                    elif choice==4:
                      display_charges()
                    elif choice==5:
                       remove_vehicle()
                    else:
                       clear_vehicles()
            else:
                if choice==1:
                  register_vehicle()
                elif choice==0:
                  print("Thanks and Good Bye!")
                  input("Please press Enter to continue...")
                else:
                  print("Invalid Input")
        else:
            print("Invalid Input")

        if choice != 0:
            input("Please press enter to continue....")

# ===================== Main =====================

if __name__ == "__main__":
    print_menu()
```
